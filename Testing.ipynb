```python
from torchvision import models
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm

def quick_test_run():
    """Quick test with VGG-16 on iml_malaria for 2 epochs"""

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Load data
    train_transforms = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    train_dataset = datasets.ImageFolder('/content/data/iml_malaria/train', transform=train_transforms)
    val_dataset = datasets.ImageFolder('/content/data/iml_malaria/val', transform=test_transforms)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

    num_classes = len(train_dataset.classes)
    print(f"Number of classes: {num_classes}")

    # Create model
    model = models.vgg16(weights=None)
    model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
    model = model.to(device)

    # Setup training
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Train for 2 epochs
    for epoch in range(2):
        model.train()
        train_loss = 0

        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

        # Validation
        model.eval()
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                val_correct += (predicted == labels).sum().item()
                val_total += labels.size(0)

        val_acc = 100 * val_correct / val_total
        print(f"Epoch {epoch+1}: Loss={train_loss/len(train_loader):.4f}, Val Acc={val_acc:.2f}%")

    print("\n✓ Test run completed successfully!")
    print("You're ready to run the full training script!")

# Run the test
quick_test_run()
```

```python
#Device setup

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

# Base hyperparameters (SAME FOR ALL RUNS for fair comparison)
BASE_CONFIG = {
    'batch_size': 32,
    'learning_rate': 0.001,
    'num_epochs': 10,  # Adjust based on time
    'input_size': 128,
}

print(f"\nBase Configuration:")
for key, value in BASE_CONFIG.items():
    print(f"  {key}: {value}")
```

```python
#Data transforms setup

# Training transforms (with augmentation)
train_transforms = transforms.Compose([
    transforms.Resize((BASE_CONFIG['input_size'], BASE_CONFIG['input_size'])),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(degrees=20),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Validation/Test transforms (no augmentation)
val_transforms = transforms.Compose([
    transforms.Resize((BASE_CONFIG['input_size'], BASE_CONFIG['input_size'])),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

print("✓ Data transforms defined")
```

```python
#Model definition

def get_vgg16_model(num_classes):
    """VGG-16 model from scratch"""
    model = models.vgg16(weights=None)
    model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
    return model

def get_mobilenet_v2_model(num_classes):
    """MobileNet-V2 model from scratch"""
    model = models.mobilenet_v2(weights=None)
    model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
    return model

print("✓ Model definition functions ready")
```

```python
#Data loading

def load_dataset(dataset_name, batch_size=32):
    """Load train, val, test datasets"""

    dataset_path = f'/content/data/{dataset_name}'

    # Load datasets
    train_dataset = datasets.ImageFolder(
        root=os.path.join(dataset_path, 'train'),
        transform=train_transforms
    )

    val_dataset = datasets.ImageFolder(
        root=os.path.join(dataset_path, 'val'),
        transform=val_transforms
    )

    test_path = os.path.join(dataset_path, 'test')
    if os.path.exists(test_path):
        test_dataset = datasets.ImageFolder(root=test_path, transform=val_transforms)
    else:
        test_dataset = val_dataset
        print(f"⚠️ No test folder, using validation set")

    # Create loaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

    num_classes = len(train_dataset.classes)
    class_names = train_dataset.classes

    print(f"\n{'='*60}")
    print(f"Dataset: {dataset_name}")
    print(f"Classes: {class_names}")
    print(f"Train: {len(train_dataset)} | Val: {len(val_dataset)} | Test: {len(test_dataset)}")
    print(f"{'='*60}\n")

    return train_loader, val_loader, test_loader, num_classes, class_names

print("✓ Data loading function ready")
```

```python
#Training model

def train_model(model, train_loader, val_loader, config, run_name):
    """Train model with checkpoint saving to Google Drive"""

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=config['learning_rate'])
    model.to(device)

    history = {
        'train_loss': [],
        'val_loss': [],
        'val_accuracy': [],
        'best_val_acc': 0.0,
        'best_epoch': 0
    }

    print(f"\n{'='*60}")
    print(f"Training: {run_name}")
    print(f"{'='*60}\n")

    for epoch in range(config['num_epochs']):
        # TRAINING
        model.train()
        running_loss = 0.0

        train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{config['num_epochs']} [Train]")
        for images, labels in train_bar:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            train_bar.set_postfix({'loss': f"{loss.item():.4f}"})

        avg_train_loss = running_loss / len(train_loader)

        # VALIDATION
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for images, labels in tqdm(val_loader, desc=f"Epoch {epoch+1} [Val]", leave=False):
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_correct += (predicted == labels).sum().item()
                val_total += labels.size(0)

        avg_val_loss = val_loss / len(val_loader)
        val_accuracy = 100.0 * val_correct / val_total

        # Update history
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['val_accuracy'].append(val_accuracy)

        # Save checkpoint (to Google Drive!)
        checkpoint_path = os.path.join(CHECKPOINT_DIR, f"{run_name}_epoch{epoch+1}.pth")
        torch.save({
            'epoch': epoch + 1,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_accuracy': val_accuracy,
        }, checkpoint_path)

        # Save best model
        if val_accuracy > history['best_val_acc']:
            history['best_val_acc'] = val_accuracy
            history['best_epoch'] = epoch + 1
            best_path = os.path.join(CHECKPOINT_DIR, f"{run_name}_BEST.pth")
            torch.save({
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'val_accuracy': val_accuracy,
            }, best_path)

        print(f"Epoch {epoch+1}: Train Loss={avg_train_loss:.4f} | Val Acc={val_accuracy:.2f}% | Best={history['best_val_acc']:.2f}%")

    return model, history

print("✓ Training function ready")
```

```python
#Evaluating

def evaluate_model(model, test_loader, class_names, run_name):
    """Evaluate model on test set"""

    model.eval()
    model.to(device)

    all_predictions = []
    all_labels = []

    print(f"\nEvaluating {run_name}...")

    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc="Testing"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(
        all_labels, all_predictions, average='weighted', zero_division=0
    )

    cm = confusion_matrix(all_labels, all_predictions)
    report = classification_report(all_labels, all_predictions,
                                   target_names=class_names, zero_division=0)

    print(f"\n{'='*60}")
    print(f"Test Results - {run_name}")
    print(f"{'='*60}")
    print(f"Accuracy:  {accuracy*100:.2f}%")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")
    print(f"\nConfusion Matrix:\n{cm}")
    print(f"{'='*60}\n")

    results = {
        'accuracy': float(accuracy),
        'precision': float(precision),
        'recall': float(recall),
        'f1_score': float(f1),
        'confusion_matrix': cm.tolist(),
    }

    return results

print("✓ Evaluation function ready")
```

```python
#Results logging

def log_results(run_name, dataset_name, model_name, run_type, config,
                history, test_results, class_names):
    """Save all results to JSON"""

    log_entry = {
        'run_name': run_name,
        'dataset_name': dataset_name,
        'model_name': model_name,
        'run_type': run_type,
        'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        'hyperparameters': config,
        'num_classes': len(class_names),
        'class_names': class_names,
        'best_val_accuracy': history['best_val_acc'],
        'best_epoch': history['best_epoch'],
        'test_accuracy': test_results['accuracy'] * 100,
        'test_precision': test_results['precision'],
        'test_recall': test_results['recall'],
        'test_f1_score': test_results['f1_score'],
        'confusion_matrix': test_results['confusion_matrix']
    }

    # Save to JSON in Google Drive
    log_path = os.path.join(RESULTS_DIR, f"{run_name}_results.json")
    with open(log_path, 'w') as f:
        json.dump(log_entry, f, indent=4)

    print(f"✓ Results saved to: {log_path}")

    return log_entry

print("✓ Logging function ready")
```
